## Q2(b) Yeild-cost Relationship

**We would employ the sklearn library**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Cost_df
```

```python
Cost_df.columns
```

**We would find the relationship between the A2+FL and the Yield**

```python
X = Cost_df[['Cost of Cultivation (`/Hectare) A2+FL']]
Y = Cost_df['Yield (Quintal/ Hectare) ']
X_train, X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
# We fit the model now
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#lets predict the test dataset 
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#We evaluate the model then
MSE = mean_squared_error(Y_test,y_pred)
r2 = r2_score(Y_test,y_pred)
```

```python
MSE
```

```python
r2
```

```python
plt.scatter(X_test,Y_test, color='b',label='Actual Data')
plt.plot(X_test,y_pred, color='r',linewidth=2, label='Regression Line')
plt.xlabel('Cost of Cultivation (`/Hectare) A2+FL', weight= 'bold')
plt.ylabel('Yeild')
plt.title('Regression Analysis: Cost of Cultivation (`/Hectare) A2+FL vs Yield')
plt.legend()
```

**From the above diagram, we can see that the is a good relationship between the two variables, Meaning that, the cost of cultivation will more than likely affect the yield amount**
